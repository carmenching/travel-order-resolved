````python
from pandas_geojson import read_geojson
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
import geopandas as gpd

stops = pd.read_csv('data_sncf/stops.txt')

region = gpd.read_file('georef-france-commune.geojson',encoding='utf-8')
stops = pd.read_csv('data_sncf/stops.txt')

communes=read_geojson('georef-france-commune.geojson')
````

Les coordonnées sont inversées pour Audun le roman

````python
lat=stops.loc[stops['stop_name']=='Audun-le-Roman-JJ.','stop_lat'][2388]
lon=stops.loc[stops['stop_name']=='Audun-le-Roman-JJ.','stop_lon'][2388]
stops.loc[stops['stop_name']=='Audun-le-Roman-JJ.','stop_lon']=lat
stops.loc[stops['stop_name']=='Audun-le-Roman-JJ.','stop_lat']=lon
````

````python
Polygon(communes['features'][0]['geometry']['coordinates'][0])

area={}
i=0
for commune in communes['features']:
    # if 1==len(commune['geometry']['coordinates'][0]):
    #     print(commune['properties']['com_name_lower'])
    coord=commune['geometry']['coordinates'][0]

    try:
        area[commune['properties']['com_name_lower']]=Polygon(coord)
    except:
        i+=1
        print(i,commune['properties']['com_name_lower'])
````

````python

def city(row):
    p=Point(row['stop_lat'],row['stop_lon'])
    for commune in area.keys():
        if p.within(area[commune]):
            return commune 
````

````python
stops['city']=stops[stops['stop_name']=="Gare de Paris-Austerlitz"].apply(city,axis=1)
````
